## Code Setup

```python
import pandas as pd
import geopandas
import os, sys
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
WINNIPEG_SHAPEFILE = os.path.join("road_network_data", "Road Network 2019-06-08 16h38", "geo_export_5373d042-8cfb-41b0-83ba-94726c9df5fe.shp")
winnipeg_map = geopandas.read_file(WINNIPEG_SHAPEFILE)
```

```python
#Prepare figure
fig, ax = plt.subplots(1, figsize=(20, 20))

#Plot Winnipeg Road Network Data
ax = winnipeg_map.plot(ax=ax)
lims = plt.axis('auto')

#Titles
plt.title('Winnipeg Road Network')
plt.ylabel('Latitude')
plt.xlabel('Longitude')

#Show the plot with a grid
plt.grid()
plt.show()
```

```python
#Winnipeg Points of Interest
#49.8951, -97.1384 - Winnipeg City Hall
#49.8844, -97.14704 - Manitoba Legislative building

df = pd.DataFrame(
 {'Point of Interest' : ['Winnipeg City Hall', 'Manitoba Legislative Building'],
 'Latitude' : [49.8951, 49.8844],
 'Longitude' : [-97.1384, -97.14704]
 })

geometry = [Point(x,y) for x, y in zip(df.Longitude, df.Latitude)]
points_of_interest_gdf = geopandas.GeoDataFrame(df, geometry=geometry)

#Prepare figure
fig, ax = plt.subplots(1, figsize=(20, 20))

#Plot Winnipeg Road Network Data
ax = winnipeg_map.plot(ax=ax)

#Plot Points of Interest
points_of_interest_gdf.plot(color='k', marker='o', ax=ax)

#Set Limits
lims = plt.axis('auto')

#Titles
plt.title('Winnipeg Road Network with some Points of Interest')
plt.ylabel('Latitude')
plt.xlabel('Longitude')

#Show the plot with a grid
plt.grid()
plt.show()
```
